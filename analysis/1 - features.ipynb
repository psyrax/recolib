## Información obtenida

La extracción de spotify consistio de los siguientes endpoints:

* [Track info](https://developer.spotify.com/documentation/web-api/reference/tracks/get-track/): Para obtener el ID de la canción y otros metadatos como album donde se incluye, etc.
* [Track features](https://developer.spotify.com/documentation/web-api/reference/tracks/get-audio-features/): Información básica musical de cada track, tiene índices diversos como la "bailabilidad", lo ruidoso y la energía de cada track.
* [Track analysis](https://developer.spotify.com/documentation/web-api/reference/tracks/get-audio-analysis/): Un monton de información que aun no entiendo bien acerca de las propiedades teorico-musicales de cada track.
